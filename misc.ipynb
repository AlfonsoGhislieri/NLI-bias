```python
from datasets import load_dataset
import nltk
import matplotlib.pyplot as plt
import numpy as np

nltk.download('punkt')  # Download the necessary tokenizer

# Load MNLI dataset
dataset = load_dataset("glue", "mnli")

# Access the hypothesis string column
hypothesis_strings = dataset["train"]["hypothesis"]

# Tokenize each hypothesis string into words and compute lengths
lengths = [len(nltk.word_tokenize(h)) for h in hypothesis_strings]

# Plot the distribution
plt.hist(lengths, bins=np.arange(30)-0.5, edgecolor='black')
plt.xlabel('Hypothesis Length (in words)')
plt.ylabel('Frequency')
plt.title('Distribution of Hypothesis Lengths')


plt.show()
```
